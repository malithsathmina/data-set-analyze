```python


# Load the dataset
file_path = "test.csv"  # Replace with your actual file path
df = pd.read_csv(file_path)

# Function to assign borough based on approximate coordinate ranges
def assign_borough(lat, lon):
    if 40.49 <= lat <= 40.92 and -74.03 <= lon <= -73.75:
        return "Manhattan"
    elif 40.55 <= lat <= 40.74 and -74.04 <= lon <= -73.85:
        return "Brooklyn"
    elif 40.57 <= lat <= 40.80 and -73.96 <= lon <= -73.70:
        return "Queens"
    elif 40.79 <= lat <= 40.91 and -73.94 <= lon <= -73.85:
        return "The Bronx"
    elif 40.50 <= lat <= 40.65 and -74.26 <= lon <= -74.05:
        return "Staten Island"
    else:
        return "Unknown"

df["Start Borough"] = df.apply(lambda row: assign_borough(row["Start Station Latitude"], row["Start Station Longitude"]), axis=1)
df["End Borough"] = df.apply(lambda row: assign_borough(row["End Station Latitude"], row["End Station Longitude"]), axis=1)

start_counts = df["Start Borough"].value_counts()
end_counts = df["End Borough"].value_counts()

total_counts = start_counts.add(end_counts, fill_value=0)

# Get the top five boroughs by total trip count
top_five_boroughs = total_counts.nlargest(7)
print(top_five_boroughs)
```

```python
birth_year_counts = df['End Station Name'].value_counts(normalize=True) * 100
birth_year_df = pd.DataFrame({'End Station Name': birth_year_counts.index, 'Percentage': birth_year_counts.values})

# Sort by most common birth years
birth_year_df = birth_year_df.sort_values(by='Percentage', ascending=False).head(20)

# Plot Table
fig, ax = plt.subplots(figsize=(6, 6))
ax.axis("tight")
ax.axis("off")
table = ax.table(cellText=birth_year_df.values,
                 colLabels=["End Station Name", "Percentage"],
                 cellLoc="center",
                 loc="center")

# Table styling
table.auto_set_font_size(False)
table.set_fontsize(10)
table.auto_set_column_width([0, 1])
table.scale(1.2, 1.2)

# Title
plt.title("End Station Name Frequency Table", fontsize=14, fontweight="bold", color="darkorange")

plt.show()
```

```python
import pandas as pd
import numpy as np

# Assuming df is your DataFrame
df['distance'] = np.sqrt(
    (df['End Station Longitude'] - df['Start Station Longitude'])**2 +
    (df['End Station Latitude'] - df['Start Station Latitude'])**2
)

# Display first few rows with distance
df[['Start Station Name', 'End Station Name', 'distance']].head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist(df['distance'], bins=1, edgecolor='black')
plt.xlabel('Distance (km)')
plt.ylabel('Frequency')
plt.title('Distribution of Ride Distances')
plt.show()
```

```python
# Round distances to one decimal place to group similar distances together
df['distance_rounded'] = df['distance'].round(3)

# Count occurrences of each distance
distance_freq = df['distance_rounded'].value_counts().reset_index()
distance_freq.columns = ['distance', 'frequency']

# Sort by distance
distance_freq = distance_freq.sort_values(by='distance')

print(distance_freq.head(10))  # Show first 10 results
```

```python
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x=distance_freq['distance'], y=distance_freq['frequency'], color='blue')
plt.xticks(rotation=45)
plt.xlabel('Distance (km)')
plt.ylabel('Frequency')
plt.title('Frequency of Ride Distances')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def generate_tag_cloud(csv_file, column_name):
    # Load dataset
    df = pd.read_csv(csv_file)
    
    # Ensure the column exists
    if column_name not in df.columns:
        print(f"Column '{column_name}' not found in dataset.")
        return
    
    # Combine all text from the column
    text = ' '.join(str(word) for word in df[column_name].dropna())
    
    # Generate word cloud
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    
    # Display the word cloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()

# Example usage
generate_tag_cloud("test.csv", "End Station ID")
```
